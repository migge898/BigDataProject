# Presentation

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import pickle

def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
allData = load_obj("allData")
```

## Visualization

```python
def visualize(bundesland, column):
    onlyBd = allData.loc[allData['bundesland'] == bundesland]
    plt.tight_layout()
    plt.title(bundesland)
    plt.plot_date(onlyBd["time_crawled"], onlyBd[column], linestyle='solid')
    plt.gcf().autofmt_xdate()
    plt.rcParams["figure.figsize"] = [14.00, 7]
    plt.show()
```

```python
for b in allData["bundesland"].unique():
    visualize(b, "todesfaelle")
```
